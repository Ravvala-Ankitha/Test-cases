```python
import pandas as pd
```

```python
data=pd.read_csv("C:/Users/admin/Downloads/12.real state.csv")
data
```

```python

# Filter for villas with prices greater than or equal to ten thousand
villas_above_10000 = data[(data['type'] == 'Residential') & (data['price'] >= 10000)]

# Group villas by city and list them
city_villa_lists = villas_above_10000.groupby('city')['street'].apply(list)

# Print the city-wise list of villas
for city, villas in city_villa_lists.items():
    print(f"City: {city}")
    print("Villas:", villas)
    print()
```

```python
# Filter for residential properties in GALT with more than 800 sq__ft
residential_above_800 = data[(data['type'] == 'Residential') & (data['sq__ft'] > 800) & (data['city'] == 'GALT')]

# Display the details of selected residential properties
selected_columns = ['street', 'sq__ft', 'sale_date', 'city']
selected_properties = residential_above_800[selected_columns]

print(selected_properties)





```

```python
# Filter for villas in CA and find the one with the lowest price
cheapest_villa = data[(data['type'] == 'Residential') & (data['state'] == 'CA')].nsmallest(1, 'price')

# Extract and print the details of the cheapest villa
cheapest_city = cheapest_villa['city'].values[0]
cheapest_street = cheapest_villa['street'].values[0]
cheapest_price = cheapest_villa['price'].values[0]

print("Cheapest Villa in CA:")
print("City:", cheapest_city)
print("Street:", cheapest_street)
print("Price:", cheapest_price)
```

```python
# Filter for residential properties that meet the criteria
filtered_properties = data[
    (data['type'] == 'Residential') &
    (data['price'] >= 60000) & (data['price'] <= 120000) &
    (data['sq__ft'] > 1450) &
    (data['beds'] >= 3) & (data['baths'] >= 2)
]

# Sort the properties by price in ascending order
sorted_properties = filtered_properties.sort_values(by='price')

# Display the top 5 residential property details
top_5_properties = sorted_properties.head(5)

print("Top 5 Residential Properties:")
print(top_5_properties)



```

```python
# Filter for residential apartments with more than 2 beds
apartments_above_2_beds = data[
    (data['type'] == 'Residential') & (data['beds'] > 2)
]

# Select and reorder columns
selected_columns = ['city', 'baths', 'sq__ft', 'price', 'type', 'beds']
selected_apartments = apartments_above_2_beds[selected_columns]

print("List of Residential Apartments with More than 2 Beds:")
print(selected_apartments)
```
