```python
import pandas as pd 
student=pd.read_csv("C:/Users/anilk/Downloads/Student.txt",delimiter='\t+',header=None,names=['Name','RollNo'])
student
```

```python
results=pd.read_csv("C:/Users/anilk/Downloads/result.txt",delimiter='\t+',header=None,names=['RollNo','Result'])
results
```

```python
combined_data=pd.merge(student,results)
combined_data
```

```python
#student names who have successfully cleared the exam
cleared_students=combined_data.loc[combined_data.Result=="pass"]
cleared_students
```

```python
#count of results
results_count=combined_data['Result'].value_counts()
print("pass Count:",results_count.get("pass"))
print("Fail Count:",results_count.get("fail"))
```

```python
#arranging student names in alphabetical order
sorted_name=combined_data.sort_values(by="Name")
sorted_name
```
